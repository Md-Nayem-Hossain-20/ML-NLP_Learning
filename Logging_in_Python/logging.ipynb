# Python Logging

Logging is a crucial aspect of any application, providing a way to track events,errors and operational information. Python's built-in logging modules offer a flexible framework for emitting log message from Python Programs.



```python
# Configure the basic logging settings

import logging

logging.basicConfig(level=logging.DEBUG)


# Log message with different severity levels
logging.debug("This is a debug message.")
logging.info("This is an info message.")
logging.warning("This is a warning message.")
logging.error("This is an error message.")
logging.critical("This is a critical message.")


```

```python
# Configuring logging
logging.basicConfig(
    filename = 'app.log',
    filemode='w',
    level = logging.DEBUG,
    format = '%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt = '%Y-%m-%d %H:%M:%S',
    force = True
)
```
