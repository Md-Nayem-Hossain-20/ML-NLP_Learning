```python
## Arithmetic operators

a = 21
b = 15

add_result = a + b
sub_result = a - b
mult_result = a * b
div_result = a / b
floor_div_result = a // b
modulus_result = a % b

exponential_result = a ** b

print(add_result)
print(sub_result)
print(mult_result)
print(div_result)
print(floor_div_result)
print(modulus_result)

print(exponential_result)
```

### Logical Operators

#### and , not, or

```python
## And operator . When both are true then only is TRUE. Rest are False.
x = True
y = True

result = x and y

x2 = True; y2 = False

result2 = x2 and y2

print(result)
print(result2)
```

```python
## Or operator . When both are False then only is False. Rest are True.
x = True
y = True

result = x or y

x2 = False; y2 = False

result2 = x2 or y2

print(result)
print(result2)
```

```python
## Not operator

x = True
not x
```
