```python
## Declaring and assigning Variables

age = 24
height = 5.5
name = "Nayem"
is_student = True

# Print the variables

print("Age: ", age)
print("Height: ", height)
print("Name: ", name)
```

```python
## Naming Convention
## Variable names should be descriptive
## They must start with a letter or '_' and contain letter,number and underscore
## Variables name are case censitive


# Valid Variable name

first_name = "Nayem"
last_name = "Nobita"

# Invalid Variable Name

# 2age = 25
# first-name = "Nayem"
# @name = "Nayem"

```

```python
# Case Sensitive

name = "Nayem"
Name = "Nobita"

name == Name
```

```python
# Python is Dynamically typed, type of a variable is determined at runtime

age = 24
height = 5.5
name = "Nayem"
is_student = True

print(type(name))

print(type(is_student))
```

```python
age = '24'

print(type(age))
```

```python
print(type(int(age)))
```

```python
# Dynamic Typing
# Python allows the type of a variable to change as the program execute

var = 10 #int
print(var , type(var))

var = "Hello"
print(var, type(var))

var = 3.14
print(var, type(var))
```

```python
## Simple Calculator

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

sum = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2


print("Sum: ", sum)
print("Difference: ", difference)
print("Product: ", product)
print("quotient: ", quotient)
```
