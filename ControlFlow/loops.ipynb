```python
## While Loop

# The while loop continues to execute as long as the condition is true.

count = 0

while count<5:
    print(count)
    count = count+1

```

```python
count = 2

while count<= 20:
    print(count)
    count = count+2
```

```python
count = 0

while count%2 == 0:
    print(count)
    count = count+1
```

```python
## loop control statement

## break

# The break statement exits the loop parmaturely

for i in range(10):
    if i==6:
        break
    print(i)
```

```python
## continue

## The continue statement skips the current iteration and continues with the next

for i in range(10):
    if i%2 == 0:
        continue
    print(i)
```

```python
## pass

## The pass statement is a null operation. It does nothing

for i in range(5):
    if i==3:
        print("The number is: ", i)
        pass
    print(i)


```

```python
##Nested loop

## a loop inside a loop

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
n = 10
sum = 0

count = 1

while count<=n:
    sum = sum+count
    count = count+1

print("Sum of the 10 natural number : ",sum)
```

```python
sum = 0

for i in range(11):
    sum = sum+i

print(sum)
```

```python
## Example - Prime Numbers between 1 and 100.

for num in range(1,101):
    if num>1:
        for i in range(2, num):
            if num%i==0:
                break
        else: 
            print(num)




```
