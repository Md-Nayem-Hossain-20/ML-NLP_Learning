### Sets

sets are a build-in data type in python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order and they do not allow duplicate elements.

```python
## Create a set

my_set = {1,2,3,4,4,5}
print(type(my_set))
print(my_set)
```

```python
my_empty_set = set()
print(my_empty_set)
```

```python
my_set = set([1,2,3,4,5])
print(my_set)
```

```python
## Basic sets operations

## Adding and removing elements

my_set.add(7)
print(my_set)
```

```python
my_set.add(6,7)
print(my_set)
```

```python
my_set.remove(3)
print(my_set)
```

```python
my_set.remove(8)
print(my_set)
```

```python
my_set.discard(11)
print(my_set)
```

```python
## Pop method

removed_element = my_set.pop()

print(removed_element)
print(my_set)
```

```python
## Clear all the elements

my_set.clear()
print(my_set)
```

```python
## Set memebership test

my_set = {1,2,3,4,5,6}

print(3 in my_set)
print(10 in my_set)
```

```python
## Mathematical operations

set1 = {1,2,3,4,5}
set2 = {2,3,4,6}

# Union

union_set = set1.union(set2)
print(union_set)
```

```python
# Intersection

intersection_set = set1.intersection(set2)

print(intersection_set)
```

```python
set1.intersection_update(set2)
print(set1)
```

```python
set1 = {1,2,3,4,6,7}
set2 = {2,5,8,9,0}

# Difference

print(set1.difference(set2))
```

```python
set2.difference(set1)
```

```python
## Symmetric difference

set1.symmetric_difference(set2)
```

```python
## Set methods 

set1 = {1,2,3,4,5,6}
set2 = {3,4,6}

## is subset
print(set1.issubset(set2))

print(set1.issuperset(set2))
```

```python
lst = [1,2,3,4,5,6]

set(lst)
```

```python
## Counting unique words in text

text = "In this tutorial we are discussing about sets."

words = text.split()

## Convert list of words to set to get unique words

unique_words = set(words)
print(unique_words)
print(len(unique_words))

```
