### Tuples

Explanation: 

Tuples are ordered collections of items that are immutable. They are similar to lists, but they are immutability makes them different.

```python
empty_tuple = ()

print(empty_tuple)
print(type(empty_tuple))
```

```python
lst = list()
print(type(lst))

tp1 = tuple()
print(type(tp1))
```

```python
numbers = tuple([1,2,3,4,5,6])

numbers
```

```python
lst = list((1,2,3,4,5,6))

lst
```

```python
mixed_tuples = (1,"Nayem",True,4.5)

print(mixed_tuples)
```

```python
## Accessing tuples elements

numbers[0]
```

```python
numbers[0:2]
```

```python
## Tuple operations

concatanation_tuple = numbers + mixed_tuples

print(concatanation_tuple)
```

```python
mixed_tuples * 3
```

```python
numbers * 3
```

```python
## Immutable nature of Tuples
## Tuples are immutable , meaning their elements can't be changed once assigned.

lst = [1,2,3,4,5]

print(lst)
```

```python
numbers[1] = "Nayem"
```

```python
## Tuple methods

print(numbers.count(1))
```

```python
print(numbers.index(3))
```

```python
## Packing and Unpacking

packed_tuple = 1,"Hello",3.14
print(packed_tuple)
```

```python
## Unpacking

a,b,c = packed_tuple

print(a)
print(b)
print(c)
```

```python
## Unpacking with *
numbers = (1,2,3,4,5,6,7)

first,*middle,last = numbers

print(first)
print(middle)
print(last)
```

```python
## Nested Tuples
## Nested List

lst = [[1,2,3],[4,5,6],[1,"Nayem",3.14]]

lst [0][2]

print(type(lst))
```

```python
lst = [[1,2,3],[4,5,6],(1,"Nayem",3.14)]

lst[2][0:3]
```

```python
nested_tuple = ((1,2,3),("a","b","c"), (True,False))

## Access the elements inside a tuple

print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item, end="")
    print()
```
